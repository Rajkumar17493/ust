# <font color=green><center>Normal Distribution</font>

* A normal distribution (Gaussian Distribution/Bell Curve) is a type of continuous probability distribution in which most data points cluster toward the middle of the range, while the rest taper off symmetrically toward either extreme.

<img src="https://mathbitsnotebook.com/Algebra2/Statistics/normalturqa.jpg">

#### Real life Examples : 

* Birthweight of Babies

* Height of People

* Shoe Sizes

* The distribution of diastolic blood pressure

#### Properties

* Mean = Median = Mode

* The total area under the curve should be equal to 1

* Symmetric

* The normal distribution curve must have only one peak. (i.e., Unimodal)

<img src="https://cdn1.byjus.com/wp-content/uploads/2019/10/normal-distribution-formula.png">

Where x is the variable, μ is the mean, σ is the standard deviation

### Standard Normal Distribution

##### z =  (x − μ) / σ

<img src="https://www.mathsisfun.com/data/images/standardizing.svg">

### Empirical Rule

<img src = "https://study.com/cimages/multimages/16/emperical2318207337824050101.png">

<img src="https://www.statology.org/wp-content/uploads/2018/10/normal_dist.png">

* About 68% of data falls within one standard deviation of the mean

* About 95% of data falls within two standard deviations of the mean

* About 99.7% of data falls within three standard deviations of the mean

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

__Example__ : Suppose the height of males at a certain school is normally distributed with mean of μ=70 inches and a standard deviation of σ = 2 inches.

Approximately what percentage of males at this school are taller than 74 inches?

<img src="https://www.statology.org/wp-content/uploads/2018/10/normaldist1.png">

```python
# calculate z score to find the location of x in terms of std from the mean
x = 74
μ = 70
σ = 2
z = (x-μ)/σ
z
```

```python
#calculate probability of a value
stats.norm.pdf(74,loc=70,scale=2)
```

```python
#calculate probability of obtaining less than a value
stats.norm.cdf(74,loc=70,scale=2)
```

```python
#calculate probability of greater than a value
1-stats.norm.cdf(74,loc=70,scale=2)
```

```python
#calculate probability for an interval
stats.norm.cdf(72,loc=70,scale=2) - stats.norm.cdf(68,loc=70,scale=2)
```

### Normal distributions with different Mean Values

```python
data=np.linspace(-4,4,100)
plt.figure(figsize=(8,6))
plt.plot(data,stats.norm.pdf(data, loc=-1, scale=1), color='blue')
plt.plot(data,stats.norm.pdf(data, loc=0, scale=1), color='green')
plt.plot(data,stats.norm.pdf(data, loc=1, scale=1), color='orange')
plt.legend(['Mean=-1','Mean=0','Mean=1'])
plt.show()
```

### Distributions with different standard deviation

```python
plt.figure(figsize=(10,6))
plt.plot(data,stats.norm.pdf(data, loc=0, scale=0.5), color='blue')
plt.plot(data,stats.norm.pdf(data, loc=0, scale=1), color='green')
plt.plot(data,stats.norm.pdf(data, loc=0, scale=1.5), color='orange')
plt.legend(['+ve Kurtosis (leptokurtic)','Normal (mesokurtic)','-ve Kurtosis (platykurtic)'])
plt.show()
```

### Kurtosis

* It's a measure of whether or not a distribution is heavy tailed or light tailed.

```python
# Kurtosis value > 3 : Leptokurtic / Heavy Tailed
# Kurtosis value = 3 : mesokurtic  / Normal
# Kurtosis value < 3 : platykurtic / Light Tailed
    
stats.kurtosis(data, fisher=False)
```

### Skewness

* It's a measure of the asymmetry of a distribution.

```python
# Negative/Left skew : Tail is on the left side, which extends towards more negative values.
# Positive/Right skew : Tail is on the right side, which extends towards more positive values.
# Zero : Distribution is Symmetric

stats.skew(data)
```

# <font color=green><center>Central Limit Theorem</font>

### Sampling Distribution of Mean

* Distribution of the sample means.

### Central Limit Theorem

* The sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough

* Regardless of whether the population has a normal, Poisson, binomial, or any other distribution, the sampling distribution of the mean will be normal.

#### Conditions :

* Sample size is sufficiently large. By convention, we consider a sample size of 30 to be “sufficiently large.”

* Sampling is random

```python
import numpy
import matplotlib.pyplot as plt
 
# sample size for each sample
sample_size = [1, 10, 50, 100] 
# list of sample means
means = []
# For each sample size,
# Generating 1, 10, 50, 100 random numbers from -40 to 40, taking their mean,
# Do the same for 1000 samples and appending it to list sample means.
for size in sample_size:
    # Generating seed so that we can get same result, every time the loop is run...
    numpy.random.seed(1)
    x = [numpy.mean(numpy.random.randint(-40, 40, size=size)) for i in range(1000)]
    means.append(x)
k = 0 
# plotting all the means in one figure
fig, ax = plt.subplots(2, 2, figsize =(8, 8))
for i in range(0, 2):
    for j in range(0, 2):
        # Histogram for each x stored in means
        ax[i, j].hist(means[k], 10, density = True)
        ax[i, j].set_title(label = sample_size[k])
        k = k + 1
plt.show()
```
