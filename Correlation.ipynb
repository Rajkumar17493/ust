# <font color=green><center>Correlation</font>

* The __measure__ of the extent to which two factors __vary together__ and of how well either factor __predicts__ the other.

* When two sets of data are strongly linked together we say they have a High Correlation.

* Correlation is Positive when the values increase together

* Correlation is Negative when one value decreases as the other increases

<img src="https://www.mathsisfun.com/data/images/correlation-examples.svg">

```python
import pandas as pd
car=pd.read_csv("C://Users//HP//Desktop//auto-mpg.csv")
car.head()
```

```python
car.corr(method='pearson')
```

```python
import seaborn as sns
import numpy as np
mask = np.triu(np.ones_like(car.corr(), dtype=bool))
sns.heatmap(data=car.corr(),annot=True,cmap='coolwarm',mask=mask)
```

```python
sns.pairplot(car,corner=True)
```
